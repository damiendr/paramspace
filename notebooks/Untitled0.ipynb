```python
from paramspace.traits_model import ModelTraits
from paramspace import choice, pchoice, uniform, normal, loguniform, quniform, sample
from traits.api import Float, Range, Any
```

```python
class TestModel(ModelTraits):
    x = Float(dist=uniform(0,1) + normal(0, 1.0))
    y = Any()
    z = Range(low=0.0, high=10.0, dist=quniform(0, 20, 0.5))
```

```python
test_space = TestModel.space(
    y = choice(
        TestModel.space(x = 1 + uniform(0, 1)),
        3 * uniform(-1, -2),
    )
) 
```
