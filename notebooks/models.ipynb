```python
import paramspace
from paramspace.traits_model import ModelTraits, load
```

```python
from traits.api import Float, Range, Any
```

```python
from paramspace.expressions import uniform
```

```python
from hyperopt import hp
```

```python
class TestModel(ModelTraits):
    x = Float(dist=uniform(0,1))
    y = Any()
```

```python
space = TestModel.space(
    y = TestModel.space(
        x = 1 + uniform(0, 1)
    )
)
```

```python
print space
```

```python
from hyperopt.pyll.stochastic import Apply, sample
```

```python
from paramspace.pyll_params import to_hp
```

```python
hpspace = to_hp(space)
```

```python
print hpspace["x"]
```

```python
print hpspace["y"]["x"]
```

```python
sample(hpspace)
```
