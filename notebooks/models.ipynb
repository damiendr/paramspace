```python
import paramspace
from paramspace.traits_model import ModelTraits
```

```python
from traits.api import Float, Range, Any
```

```python
from scipy.stats import uniform
```

```python
from hyperopt import hp
```

```python
class TestModel(ModelTraits):
    x = Float(dist=uniform(0,1))
    y = Any()
```

```python
space = TestModel.space(
    y = TestModel.space()
)
```

```python
space
```

```python
from hyperopt.pyll.stochastic import Apply, sample
sample(space)
```

```python
x = space['x']
```

```python
space['x'] = Apply(x.name, x.pos_args, x.named_args)
```
